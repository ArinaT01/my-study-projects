## Data Preprocessing

### Handling Missing Values

**Display the number of missing values for each column.**

```python
data.isna().sum()
```

**Visualize the percentage of missing values by column:**

```python
def pass_value_barh(df):
    try:
        # Calculate the percentage of missing values and create the plot
        missing_percentage = df.isna().mean() * 100
        missing_df = missing_percentage.to_frame().rename(columns={0: 'space'})
        missing_df = missing_df.query('space > 0').sort_values(by='space', ascending=True)
        
        # Create the horizontal bar plot
        missing_df.plot(kind='barh', figsize=(19, 6), rot=0, legend=False, fontsize=12, color='steelblue')

        # Set the title and labels for the axes
        plt.title('Percentage of Missing Values by Column\n', fontsize=22, color='SteelBlue')
        plt.xlabel('Percentage of Missing Values', fontsize=14)
        plt.ylabel('Column Name', fontsize=14)
        plt.xlim(0, 100)  # Set the x-axis scale from 0 to 100

        # Display the plot
        plt.show()
    except Exception as e:
        print(f'Error occurred: {e}')

# Example usage of the function
pass_value_barh(data)
```

**Fill missing values in certain columns with zeroes where it's logical.**

```python
data['balcony'] = data['balcony'].fillna(0) 
data['floor'] = data['floor'].fillna(0) 
data['kitchen_area'] = data['kitchen_area'].fillna(0)  
data['parks_around3000'] = data['parks_around3000'].fillna(0) 
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)
data['cityCenters_nearest'] = data['cityCenters_nearest'].fillna(0)
data['is_apartment'] = data['is_apartment'].fillna('False')
```

**Fill missing values in `ceiling_height` based on the median for each locality.**

```python
median_ceiling_height = data.groupby('locality_name')['ceiling_height'].median()
data['ceiling_height'] = data['ceiling_height'].fillna(data['locality_name'].map(median_ceiling_height))
```

There are many gaps and outliers in the ceiling_height column, the data with missing values looks normal, no patterns are visible, so the decision was made to fill the gaps with the median value.

**Remove rows with missing values in `floors_total` column, since the quality of the solution to the main problem will not suffer from this.**

```python
data = data.dropna(subset=['floors_total'])
data['floors_total'] = data[['floors_total', 'floor']].max(axis=1)
```

**Fill missing values for distance-related columns using the median for each locality.**

```python
columns_to_fill = ['parks_nearest', 'airports_nearest', 'ponds_nearest', 'cityCenters_nearest']
for column in columns_to_fill:
    median_distance = data.groupby('locality_name')[column].transform('median')
    data[column] = data[column].fillna(median_distance)
```

**Fill missing values in `living_area` based on the median for the corresponding number of rooms.**

```python
median_living_area = data.groupby('rooms')['living_area'].median()
data['living_area'] = data['living_area'].fillna(data['rooms'].map(median_living_area))
```

```python
# Ensure that 90% of total area is greater than the sum of living and kitchen areas
incorrect_areas_total = data.query('total_area * 0.9 < (living_area + kitchen_area)')
if not incorrect_areas_total.empty:
    print("Incorrect values for total_area:")
    print(incorrect_areas_total[['total_area', 'living_area', 'kitchen_area']])

# Ensure that living area is not greater than total area
incorrect_areas_living = data.query('living_area >= total_area')
if not incorrect_areas_living.empty:
    print("\nIncorrect values for living_area:")
    print(incorrect_areas_living[['total_area', 'living_area', 'kitchen_area']])

# Ensure that kitchen area is not greater than living area or total area
incorrect_areas_kitchen = data.query('kitchen_area >= total_area | kitchen_area >= living_area')
if not incorrect_areas_kitchen.empty:
    print("\nIncorrect values for kitchen_area:")
    print(incorrect_areas_kitchen[['total_area', 'living_area', 'kitchen_area']])
    
# Ensure that the number of floors in a building is not less than the floor of the apartment
incorrect_floor = data.query('floors_total < floor', engine='python')
if not incorrect_floor.empty:
    print("\nIncorrect values for floors:")
    print(incorrect_areas_kitchen[['floor', 'floors_total']])
```

```python
print(data.isna().sum()) # Final check on missing values
```

## Add New Columns

**Price per square meter.**

```python
# Calculate the price per square meter by dividing the last price by the total area, and round to 2 decimal places.
data['price_per_square_meter'] = (data['last_price'] / data['total_area']).round(2) 
```

**Day of week of announcement publication.**

```python
# Extract the day of the week from the 'first_day_exposition' column (0 = Monday, 1 = Tuesday, etc.).
data['day_of_week'] = data['first_day_exposition'].dt.dayofweek
```

**Month of week of announcement publication.**

```python
# Extract the month from the 'first_day_exposition' column.
data['month'] = data['first_day_exposition'].dt.month
```

**Year of week of announcement publication.**

```python
# Extract the year from the 'first_day_exposition' column.
data['year'] = data['first_day_exposition'].dt.year 
```

**Type of apartment floor.**

```python
# Determine the floor type based on 'floor' and 'floors_total':
for index, row in data.iterrows():
    if row['floor'] <= 0 or row['floors_total'] <= 0:
        data.at[index, 'floor_type'] = 'undefined'
    elif row['floor'] == 1:
        data.at[index, 'floor_type'] = 'first'
    elif row['floor'] == row['floors_total']:
        data.at[index, 'floor_type'] = 'last'
    else:
        data.at[index, 'floor_type'] = 'other'
```

**Distance to city center in kilometers.**

```python
# Rename the column from 'cityCenters_nearest' to 'city_centers_nearest'.
data.rename(columns={'cityCenters_nearest': 'city_centers_nearest'}, inplace=True)
```

```python
# Convert the distance to the city center from meters to kilometers, round to the nearest whole number.
data['city_centers_nearest_km'] = (data['city_centers_nearest'] / 1000).round()
```

**Conclusion**

New columns have been added to the dataset: `price_per_square_meter` for cost per square meter, `day_of_week, month, and year` for publication date details, `floor_type` to categorize the floor level as 'first', 'last', 'other', or 'undefined', and `city_centers_nearest_km` for distance to the city center in kilometers. These updates provide more detailed and useful features for analysis.

### Analyzing and Filtering Data

```python
# Function to plot histograms on one large figure
def plot_all_histograms(df, columns, bins):
    n_cols = 2  # Number of columns in the grid
    n_rows = (len(columns) + 1) // n_cols  # Number of rows based on number of columns to plot
    
    plt.figure(figsize=(15, n_rows * 5))  # Adjust the figure size to fit all plots
    for i, column in enumerate(columns):
        plt.subplot(n_rows, n_cols, i + 1)
        df[column].hist(bins=bins)
        plt.title(column.replace('_', ' ').title())
        plt.xlabel(column.replace('_', ' ').title())
        plt.ylabel('Frequency')
    plt.tight_layout()  # Adjust subplots to fit into figure area
    plt.show()

# Parameters to correct with min and max thresholds
parameters = {
    'total_area': (15, 300),
    'living_area': (5, 250),
    'kitchen_area': (0, 50),
    'last_price': (100_000, 100_000_000),
    'rooms': (1, 10),
    'ceiling_height': (2, 5),
    'city_centers_nearest': (0, 40_000)
}

# Create a copy of the data
filtered_data = data.copy()

# Function to filter out anomalies and correct values
def correct_anomalies(df, column, min_val, max_val):
    df[column] = df[column].apply(lambda x: x if min_val <= x <= max_val else None)
    return df.dropna(subset=[column])

# Apply filtering for each parameter
for column, (min_val, max_val) in parameters.items():
    filtered_data = correct_anomalies(filtered_data, column, min_val, max_val)

# Plot histograms for all columns on one figure
plot_all_histograms(filtered_data, list(parameters.keys()), bins=30)

# Specific plot for floor type (categorical)
plot_all_histograms(filtered_data, ['floor_type'], bins=30)
```

### Identify the factors that most influence the overall (total) cost of the property.

```python
# Creating a pivot table
numeric_columns = ['total_area', 'living_area', 'kitchen_area', 'rooms', 'days_exposition', 'day_of_week', 'month', 'year']
pivot_table = filtered_data.pivot_table(index='last_price', 
                                        values=numeric_columns, 
                                        aggfunc='median')

# Plotting graphs
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(12, 12))

# Graphs showing price dependence on various parameters
sns.scatterplot(data=filtered_data, x='total_area', y='last_price', ax=axes[0, 0])
axes[0, 0].set_title('Total Area')

sns.scatterplot(data=filtered_data, x='living_area', y='last_price', ax=axes[0, 1])
axes[0, 1].set_title('Living Area')

sns.scatterplot(data=filtered_data, x='kitchen_area', y='last_price', ax=axes[0, 2])
axes[0, 2].set_title('Kitchen Area')

sns.scatterplot(data=filtered_data, x='rooms', y='last_price', ax=axes[1, 0])
axes[1, 0].set_title('Number of Rooms')

sns.scatterplot(data=filtered_data, x='days_exposition', y='last_price', ax=axes[1, 1])
axes[1, 1].set_title('Days Listed')

sns.scatterplot(data=filtered_data, x='floor_type', y='last_price', ax=axes[1, 2])
axes[1, 2].set_title('Floor Type')

sns.scatterplot(data=filtered_data, x='day_of_week', y='last_price', ax=axes[2, 0])
axes[2, 0].set_title('Day of the Week Listed')

sns.scatterplot(data=filtered_data, x='month', y='last_price', ax=axes[2, 1])
axes[2, 1].set_title('Month Listed')

sns.scatterplot(data=filtered_data, x='year', y='last_price', ax=axes[2, 2])
axes[2, 2].set_title('Year Listed')

# Arranging the layout of the graphs
plt.tight_layout()
plt.show()
```

```python
# Building a correlation matrix
correlation_matrix = pivot_table.corr()

# Creating a heatmap for the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1, vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

**Conclusion**

1. **Strong positive correlations:**
- `Total_area` and `living_area` have a high correlation (0.921), which is logical since living area constitutes a significant portion of the total area.
- Number of `room` is also strongly correlated with `living_area` (0.888) and `total_area` (0.837), which is expected—more rooms usually mean a larger area.

2. **Moderate positive correlations:**
- `Kitchen_area` has a moderate correlation with `total_area` (0.526) and `living_area` (0.440), as the kitchen takes up part of the total area.
- `Days_of_exposition` moderately correlate with `total_area` (0.200), which may suggest that larger properties take longer to sell.

3. **Strong negative correlations:**
- `Year` and `days_of_exposition` have a significant negative correlation (-0.608), which might indicate that over time, the selling time for properties has decreased.

4. **Weak correlations:**
- The remaining correlations are quite weak, indicating that they have a lesser impact on the parameters. For instance, `day_of_week` and other parameters show almost no significant correlation.

Overall, the data shows logical relationships between areas, the number of rooms, and the time properties spend on the market.

To supplement the conclusions about price trends, it's important to consider not only the average price but also other significant factors, such as the number of listings and the average apartment area over the years. This will help identify the relationship between price changes and apartment size, as well as assess the impact of the number of listings on the accuracy of the analysis.

```python
# Create pivot tables
pivot_table_price = filtered_data.pivot_table(index='year', values='last_price', aggfunc='mean')
pivot_table_area = filtered_data.pivot_table(index='year', values='total_area', aggfunc='mean')
pivot_table_count = filtered_data.pivot_table(index='year', values='last_price', aggfunc='count')

# Plot line charts
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot average price change by year
color = 'tab:red'
ax1.set_xlabel('Year')
ax1.set_ylabel('Average Price', color=color)
ax1.plot(pivot_table_price.index, pivot_table_price['last_price'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Create a second axis for average area
ax2 = ax1.twinx()  
color = 'tab:blue'
ax2.set_ylabel('Average Area', color=color)  
ax2.plot(pivot_table_area.index, pivot_table_area['total_area'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Create a third axis for the number of listings
ax3 = ax1.twinx()
color = 'tab:green'
ax3.spines['right'].set_position(('outward', 60))  
ax3.set_ylabel('Number of Listings', color=color)
ax3.plot(pivot_table_count.index, pivot_table_count['last_price'], color=color)
ax3.tick_params(axis='y', labelcolor=color)

plt.title('Changes in Average Price, Area, and Number of Listings by Year')
plt.show()
```

**Conclusion**
- **Change in the number of records.** A small number of listings in certain years can significantly skew the analysis results. It’s important to ensure that there is enough data for objective conclusions.

- **Change in apartment size.** From 2017-2018, there is a drop in the average apartment size, likely due to the entry of mass sellers into the market. This could have led to a decrease in the average apartment price during these years, rather than an actual decline in market prices.

- **Distribution of listings by location.** Average prices can vary significantly depending on the location (for example, apartments in St. Petersburg vs. Vyborg). Uneven distribution of locations across different years can also distort the overall picture.

- **Completeness of data by year.** It's important to account for seasonal price fluctuations (e.g., summer price reductions). Incomplete data for 2014 and 2019 may affect the analysis results, as there could be missing data for certain months.

Thus, to draw accurate conclusions, it’s necessary to consider the combined effect of all these factors.
